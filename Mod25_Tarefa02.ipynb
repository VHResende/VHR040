1- Calcule de forma numérica a integral da função $x^3$ avaliada entre os pontos $1$ e $2$. Identifique o número de retângulos que você usou.  
2- Calcule essa área de forma analítica.

1. **Cálculo Numérico da Integral** <br>
A abordagem numérica, aqui, é chamada de método dos retângulos (ou regra do retângulo), que é uma forma simples de calcular uma integral aproximada.

2. **Cálculo Analítico da Integral** <br>
Para calcular a integral analiticamente, usamos as regras de integração conhecidas, confrome os códigos abaixo.

**Visualizando no Código**:

**Cálculo Numérico com o Método dos Retângulos**:

```python
# Cálculo numérico da integral usando o método dos retângulos

# Definindo a função f(x) = x^3
def f(x):
    return x**3

# Definindo os parâmetros
a = 1  # limite inferior
b = 2  # limite superior
n = 1000  # número de retângulos (quanto maior o n, maior a precisão)

# Calculando a largura de cada retângulo
dx = (b - a) / n

# Calculando a área aproximada (somatório das áreas dos retângulos)
area_numerica = sum(f(a + i * dx) * dx for i in range(n))

print("Área Numérica Aproximada:", area_numerica)
```

**Cálculo Analítico**:

```python
# Integral analítica
def integral_analitica(x):
    return (x**4) / 4

# Cálculo da integral analiticamente entre os pontos 1 e 2
area_analitica = integral_analitica(b) - integral_analitica(a)

print("Área Calculada Analiticamente:", area_analitica)
```

**No gráfico**:

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função f(x) = x^3
def f(x):
    return x**3

# Definindo os parâmetros
a = 1  # limite inferior
b = 2  # limite superior
n = 10  # número de retângulos para visualização

# Valores de x para a função analítica
x_vals = np.linspace(a, b, 1000)
y_vals = f(x_vals)

# Calculando os retângulos
x_rect = np.linspace(a, b, n+1)[:-1]
y_rect = f(x_rect)
dx = (b - a) / n

# Plotando a função e os retângulos
plt.figure(figsize=(8,6))

# Plotando a função original
plt.plot(x_vals, y_vals, label="f(x) = x³", color='blue')

# Plotando os retângulos
plt.bar(x_rect, y_rect, width=dx, align='edge', edgecolor='black', color='lightgray', alpha=0.7, label="Retângulos (método numérico)")

# Informações do gráfico
plt.title("Cálculo Numérico da Integral usando Retângulos")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
